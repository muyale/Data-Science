# INDIVIDUALS AND THEIR AWARDS

```python
# Individuals and their awArds
df[df['Full Name']=='Albert Einstein']
```

```python
def find_winner(name,df=df):
    print(f"Birthday is:{df[df['Full Name']==name]['Birth Date']}")
    print(f"Birth city is:{df[df['Full Name']==name]['Birth City']}")
    print(f"The year of award is :{df[df['Full Name']==name]['Year']}")
    print(f"The category is :{df[df['Full Name']==name]['Category']}")
    print(f"The motivation was  :{ df[df['Full Name']==name]['Motivation']}")
    print(f"The organization name is: {df[df['Full Name']==name]['Organization Name']}")
```

```python
find_winner('Albert Einstein')
```

```python
find_winner('Wangari Muta Maathai')
```
