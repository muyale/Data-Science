# OBJECT ORIENTED PROGRAMMING
OBJECT ORIENTED PROGRAMMING(OOP) is a paradgim that uses objects and classes to model real world entities.It does so by implementing
real world entitities like inheritance in the program.The main  concept in OOP is to bind data and the functions that work 
together as single units . This method structures programs by bundling related properties and behaviors into individual objects .
Objects could represent a person with properties such as name and age with behaviors such as walking running or talking .OOP then
brings all the attributes ,behaviours to match real life . It is the basis of real world modelling using computer programming

A class is a collection of objects. A class contains the blueprints or the prototype from which the objects are being created. It is a logical entity that contains some attributes and methods.

# Attributes and the __init__ method 
Attributes are behaviours that instances in our class can have .Attributes created in the init are called instance attributes .The values are specific to instances 

```python
# we use __init__ to declare which attributes we require 
class Person:
    def __init__(self,name,age)
    """__init__ method sets the initial state of the object  by assigning the values of the object properties,
    Init fills out the self parameter automatically leaving us with the other parameters to fill """
    
    
    self.name = name 
    self.age = age 
    """Name and age are the attributes that instances of our Person class can have """
```

# Instatiating a class 
When you instatiate a class in Python ,Python creates a new instance of the class and passes it to the first parameter of the init method so you dont have to fill the self parameter 

```python
# lets create a Person class with running instance 
class Person :
    def __init__(self,name,age):
        self.name = name
        self.age  = age 
    def running(self): # instance
        return (f"I am {self.name} and I am  running")
    def description(self):
        return (f"I am {self.name} and I am {self.age}yrs old")
    
```

```python
edgar = Person("Edgar",23)
# You can access the class attributes using the dot notation
print(edgar.description())
print("\n")
print(edgar.running())
```

# Inheritance in Python classes 
Inheritance is the process by which a child class takes the attributes and methods of a Parent class .

```python
# Example of inheritance 
class Person: # parent child
    species_name = "Homo Sapiens" # This is a class attributes ,usually defined outside init 
class Male(Person):
    def __init__(self,name):
        self.name = name
    
```

```python
male = Male("Edgar")
male.species_name  # The male class has inherited from the Parent class 
```

```python
# Example of a class
# Ill use OOP to describe a shirt ,giving it attributes and creating methods to print them out 
# The attributes include shirt color ,shirt size,shirt designer and price 
# The methods include double price ,triple price ,functions to print out the info on our shirts
class Shirt():
    def __init__(self,color,size,price,designer):
        self.color = color
        self.size = size
        self.price = price
        self.designer = designer
    def  shirt_color(self):
        color = self.color
        return (f"The color of the shirt is {self.color}")
    def shirt_price(self,price):
        return self.price
    def shirt_double_price(self):
        self.price = 2*self.price
        return self.price
    def shirt_designer(self):
        return(f"The Designer of the shirt is { self.designer}")
    def shirt_size(self):
        return(f"The Designer of the shirt is { self.designer}")
```

```python
# Example 2 :Suppose I was to design an Employee base with all their attributes ,How will I use OOP ?
class Employee():
    def __init__(self,name,department,salary,sex,time_in,time_out):
        self.name= name
        self.department = department
        self.salary = salary
        self.sex = sex
        self.time_in =time_in
        self.time_out = time_out 
    def employee_name(self):
        return (f"My name is {self.name}")
    def employee_sex(self):
        return (f"Employee sex is {self.sex}")
    def employee_department(self):
        return (f"Employee Department is  {self.department}")
    def employee_total_working(self):
        total_working_hours = self.time_out-self.time_in
        return (f"The employee works for a total of {total_working_hours }Hours")
```

# Histogram 
 For this plot ,we are required to fill in the data, x  and an optionally hue ,Hue is like a filter 
 Example sns.histplot(x=?,data=?,hue=None)

```python

# lets plot a histogram to show the distribution of  age
 
fig = plt.figure(figsize=(20,8))
sns.histplot(x='age',data=df)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram showing the distribution of Age")
plt.show() # To create more histogram we will have to rewrite this code but we can write a function to do this 
```

```python
def plot_histogram(x,df=df):
    """Args: X is the variable which we wish to study ,everytime we call the function we will be required to fill x.
    The function creates a figure or canvas and does the filling automatically .
    Functions are reusable and so make our work easier
    Returns :
    Plotted histogram """
    fig = plt.figure(figsize=(20,8))
    sns.histplot(x=x,data=df)
    plt.xlabel(f"{x}")
    plt.ylabel("Frequency")
    plt.title(f"Histogram showing the distribution of {x}")
    plt.show()
```

```python
# example 
plot_histogram(x='age')
```

```python
# I will create a function that returns a barplot should we need to create more than one at any point in our project
def plot_barplot(x,y,df=df):
    """Args: 
    X is the  independent variable which we wish to study and Y is the dependent variable ,
    everytime we call the function we will be required to fill x and y 
    The function creates a figure or canvas fills x and y .
    Functions are reusable and so make our work easier
    Returns :
        Plotted histogram """
    fig = plt.figure(figsize=(20,8))
    sns.barplot(x=x,y=y,data=df)
    plt.xlabel(f"{x}")
    plt.ylabel(f"{y}")
    plt.title(f"Barplot  showing the distribution of {x} and {y}")
    plt.show()
```

# Countplot
Countplots are good for categorical data that is univariate and the hues make our plots interesting
For this plot ,we are required to fill in the data, x  and an optionally hue ,Hue is like a filter 
 Example sns.histplot(x=?,data=?,hue=None)

```python
fig = plt.figure(figsize=(20,8))
sns.countplot(x='sex',data=df)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Countplot showing the distribution of Age")
plt.show()
```

```python
def plot_countplot(x,hue=None,df=df):
    """Args: 
    X is the categorical variable that we wish to study.The hue is a filter on which we can consider for this plot.
    Returns :
        Plotted Countplot """
    fig = plt.figure(figsize=(20,8))
    sns.countplot(x=x,hue=hue,data=df)
    plt.xlabel(f"{x}")
    plt.ylabel("Total Number")
    plt.title(f"Countplot  showing the distribution of {x} ")
    plt.show()
```

```python
plot_countplot(x="smoker",hue='sex')
```

# PART 3 : BRINGING IT ALL TOGETHER : OOP IN DATASCIENCE

```python
class DataSciencePlots():
    def __init__(self,df):
        self.df =df
    def plot_histogram(self,x,df=df):
        """Args: X is the variable which we wish to study ,everytime we call the function we will be required to fill x.
        The function creates a figure or canvas and does the filling automatically .
        Functions are reusable and so make our work easier
        Returns :
        Plotted histogram """
        fig = plt.figure(figsize=(20,8))
        sns.histplot(x=x,data=df)
        plt.xlabel(f"{x}")
        plt.ylabel("Frequency")
        plt.title(f"Histogram showing the distribution of {x}")
        plt.show()
    def plot_barplot(self,x,y,df=df):
        """Args: 
        X is the  independent variable which we wish to study and Y is the dependent variable ,
        everytime we call the function we will be required to fill x and y 
        The function creates a figure or canvas fills x and y .
        Functions are reusable and so make our work easier
        Returns :
            Plotted histogram """
        fig = plt.figure(figsize=(20,8))
        sns.barplot(x=x,y=y,data=df)
        plt.xlabel(f"{x}")
        plt.ylabel(f"{y}")
        plt.title(f"Barplot  showing the distribution of {x} and {y}")
        plt.show()
    def plot_countplot(self,x,hue=None,df=df):
        """Args: 
        X is the categorical variable that we wish to study.The hue is a filter on which we can consider for this plot.
        Returns :
            Plotted Countplot """
        fig = plt.figure(figsize=(20,8))
        sns.countplot(x=x,hue=hue,data=df)
        plt.xlabel(f"{x}")
        plt.ylabel("Total Number")
        plt.title(f"Countplot  showing the distribution of {x} ")
        plt.show()
    
        
```

```python
plotter = DataSciencePlots(df=df)
```

```python
plotter.plot_histogram(x="charges")
```

```python
plotter.plot_countplot("sex")
```
