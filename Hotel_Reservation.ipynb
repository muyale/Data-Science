```python
# A function that returns column analysis values
def column_analysis(column,df=df):
    """This function returns unique values,frequent elements,statistical analysis and info about columns
    Args:columns(str): the column we intend to study
    Returns:
    """
    print(f"The total number of unique values is {df[column].nunique()}")
    print(f"The most frequent elements are : {df[column].value_counts().head(10)}")
    print(f"The least frequent elements are : {df[column].value_counts().tail(5)}")
    print(f"The statistical measures are :{df[column].describe()}")
```

```python
df.columns
df.drop(['Booking_ID'],axis=1,inplace=True)
```

```python
X= df.drop(['booking_status'],axis=1)
y=df['booking_status']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=600)
model.fit(X_train,y_train)
prediction = model.predict(X_test)
print(prediction)
sns.countplot(x=prediction)
plt.title('A countplot for the  Random Forest predictions')
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,prediction))
print(confusion_matrix(y_test,prediction))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
dpredict=dtree.predict(X_test)
print(dpredict)
```

```python
sns.countplot(x=dpredict)
plt.title('A countplot for the  Decision tree predictions')
```

```python
print('The following is  A Classification report for Decision Tree Model')
print(classification_report(y_test,dpredict))
print(confusion_matrix(y_test,dpredict))
```
