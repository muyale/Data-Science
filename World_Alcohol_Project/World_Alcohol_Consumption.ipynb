```python
def generalized_analysis(column,df=usa_df):
    print(f"The Unique Values in our {column} are {df[column].nunique()}")
    print("\n")
    print(f"The Most appearing elements  in our {column} are {df[column].value_counts().head()}")
```

```python
for column in usa_df.columns:
    generalized_analysis(column=column)
```

```python
total_beer = usa_df['Beer (Per capita consumption)'].sum()/usa_df['All beverages (Per capita consumption)'].sum()
total_wine = usa_df['Wine (Per capita consumption)'].sum()/usa_df['All beverages (Per capita consumption)'].sum()
total_spirits = usa_df['Spirits (Per capita consumption)'].sum()/usa_df['All beverages (Per capita consumption)'].sum()
total_beer = total_beer *100
total_wine = total_wine*100
total_spirits = total_spirits*100
listed = (total_beer,total_wine,total_spirits)
labels = ['Beer','Wine','Spirits']
plt.pie(x=listed,labels=labels,startangle=105)
plt.title('Beer Spirits and Alcohol Consumption')
```

```python
import datetime
world_df['Date'] = pd.to_datetime(world_df['Date'])
world_df.head()
```

```python
world_df.nunique()
```

```python
"""Total number of females consuming alcohol"""
males = world_df[world_df['Gender']=='Male']['Count'].sum()
females = world_df[world_df['Gender']=='Female']['Count'].sum()
total = males + females
gender = (males,females)
names = ['Male','Female']
print(f"The total number of Men drinking Alcohol is :{males} and percentage is{(males/total)*100}")
print(f"The total number of Women drinking Alcohol is:{females} and percentage is{(females/total)*100}")
# The Gender that consumes the most alcohol is The Women ,worldwide:the two genders are not so far off
fig = plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.pie(x=gender,labels=names)
plt.title('Gender Distribution Pie Chart')
plt.subplot(2,2,2)
sns.countplot(x=world_df['Gender'])
plt.title('A countplot for Gender Distribution')
plt.subplot(2,2,3)
sns.histplot(x=world_df['Gender'])
plt.title('A hisplot for Gender Distribution')
```

```python
"""Alcohol consumption by Gender ,Over the years ,Which gender has drunk more """
males = world_df[world_df['Gender']=='Male']['Count']
year_m = world_df[world_df['Gender']=='Male']['Date']
females = world_df[world_df['Gender']=='Female']['Count']
year_f = world_df[world_df['Gender']=='Female']['Date']
fig= plt.figure(figsize=(15,10))
fig.suptitle('Comparing Male and Female alcohol consumption over time')
plt.subplot(2,2,1)
sns.lineplot(x=year_m,y=males)
plt.title('Male Alcohol Consumption over the years')
plt.subplot(2,2,2)
sns.lineplot(x=year_f,y=females)
plt.title('Female Alcohol Consumption over the years')
```

```python
"""Analysis Country Alcohol consumption ,I will focus on UK,USA and UK """
country = world_df[world_df['Countries']=='Kenya']['Count']
date= world_df[world_df['Countries']=='Kenya']['Date']
fig = plt.figure(figsize=(15,10))
fig.suptitle('Alcohol Consumption analysis by countries')
plt.subplot(2,2,1)
sns.lineplot(x= world_df[world_df['Countries']=='Kenya']['Date'],y= world_df[world_df['Countries']=='Kenya']['Count'])
plt.title('Alcohol Consumption in Kenya')
plt.subplot(2,2,2)
sns.lineplot(x= world_df[world_df['Countries']=='United Kingdom']['Date'],y= world_df[world_df['Countries']=='United Kingdom']['Count'])
plt.title('Alcohol Consumption in UK')
plt.subplot(2,2,3)
sns.lineplot(x= world_df[world_df['Countries']=='United States']['Date'],y= world_df[world_df['Countries']=='United States']['Count'])
plt.title('Alcohol Consumption in USA')
```

```python
top_ten_countries = world_df.groupby('Countries')['Count'].sum().sort_values(ascending=True).tail(10)
# To get the countries with the most number of alcohol users
# China Leads ,because of its high population?
top_ten = (134107,146595,182036,190825,205345,233493,342678,345774,652428,1176094)
countries = ['United States','Sweden','France','Portugal','Poland','Brazil','Russia','Phillipines','Indonesia','China']
plt.pie(x=top_ten,labels = countries)
plt.title('Top Ten Countries interms of Alcohol consumption')
```
