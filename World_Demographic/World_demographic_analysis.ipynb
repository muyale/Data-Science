```python
df.dropna(inplace=True)
```

```python
df['Adults (ages 15+) and children (ages 0-14) newly infected with HIV'] = df['Adults (ages 15+) and children (ages 0-14) newly infected with HIV'].fillna(df['Adults (ages 15+) and children (ages 0-14) newly infected with HIV'].mean())
df['Adults (ages 15-49) newly infected with HIV'] = df['Adults (ages 15-49) newly infected with HIV'].fillna(df['Adults (ages 15-49) newly infected with HIV'].mean())
```

```python
df.isnull().sum()
```

```python
df['Ratio of female to male labor force participation rate (%) (national estimate)'] =df['Ratio of female to male labor force participation rate (%) (national estimate)'].fillna(df['Ratio of female to male labor force participation rate (%) (national estimate)'].mean())
```

```python
df.isnull().sum()
```

```python
df['Total reserves (includes gold, current US$)'] = df['Total reserves (includes gold, current US$)'].fillna(df['Total reserves (includes gold, current US$)'].min())
```
