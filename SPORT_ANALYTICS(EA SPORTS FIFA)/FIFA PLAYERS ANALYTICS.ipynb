# Overall Ratings :: General Analysis

```python
fig = plt.figure(figsize=(20,8))
plt.subplot(2,2,1)
sns.histplot(x='overall',data=df,color='black')
fig.suptitle("The Overall ratings distributions")
plt.subplot(2,2,2)
sns.boxplot(x='overall',data=df,color='black')
fig.suptitle("The Overall ratings distributions :Boxplot and Histplot")

plt.show()
```

```python
fig = plt.figure(figsize=(20,8))
plt.subplot(2,2,1)
sns.histplot(x='height_cm',data=df,color='black')
fig.suptitle("The Overall ratings distributions")
plt.subplot(2,2,2)
sns.histplot(x='pace',data=df,color='yellow')
fig.suptitle("The Overall ratings distributions : Height and Pace ")
plt.subplot(2,2,3)
sns.jointplot(y='height_cm',x ='pace',data=df,markers="*")
plt.title("Does Height Affect the Pace ,are shorter players pacier?") # Yes ,Pace and Height have a relationship
plt.show()
```

```python
# PASSING SHOOTING AND ATTACKING, OVERALL RATINGS
attacking_traits = ['attacking_crossing', 'attacking_finishing',
       'attacking_heading_accuracy', 'attacking_short_passing',
       'attacking_volleys']
skill_traits = ['skill_dribbling', 'skill_curve',
       'skill_fk_accuracy', 'skill_long_passing', 'skill_ball_control']
movement_traits = ['movement_acceleration', 'movement_sprint_speed', 'movement_agility',
       'movement_reactions', 'movement_balance']
power_traits = ['power_shot_power',
       'power_jumping', 'power_stamina', 'power_strength', 'power_long_shots']
```

```python
# I will create a function that plots a histogram and use it to plot 
def plot_history(column,df=df.head(50)):
    plt.figure(figsize=(20,8))
    sns.histplot(x=column,data=df)
    plt.xlabel(f"{column}")
    plt.ylabel("Total of Each Column")
    plt.title(f"The Overall distribution of {column} column")
```

# ATTACKING TRAITS


```python
# First we look at the first 50 players' attacking traits
for trait in attacking_traits : plot_history(trait)
```

# SKILL TRAITS 

```python
# Next we look at the first 50 players' skill traits
for trait in skill_traits : plot_history(trait)
```

# MOVEMENT TRAITS

```python
# Then we look at the first 50 players' movement traits
for trait in movement_traits : plot_history(trait)
```

# POWER TRAITS
