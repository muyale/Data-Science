```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
df = pd.read_csv("C:\\Datasets\\billboard\\charts.csv")
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.drop(['last-week'],axis=1,inplace=True)
```

```python
#Most appearing artists on The charts
df['artist'].value_counts().head(20)
# Taylor swift has the most number of chart appearance
```

# STUDYING MADONNA ,TAYLOR SWIFT AND ELTON JOHN:

```python
plt.figure(figsize=(20,8))
sns.lineplot(y= df[df['artist']=='Taylor Swift']['rank'],x= df[df['artist']=='Taylor Swift']['date'])
plt.title('The Rise of Taylor Swift interms of Rank')
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(y= df[df['artist']=='Elton John']['rank'],x= df[df['artist']=='Elton John']['date'])
plt.title('The Rise of Elton John interms of Rank')
# Between 1972 and 1976 Elton John was almost unstopabble
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(y= df[df['artist']=='Madonna']['rank'],x= df[df['artist']=='Madonna']['date'])
plt.title('Maddona Golden years interms of Rank')
# Madonna Dominated the 80 s and early 21 st century
```
