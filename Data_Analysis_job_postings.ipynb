```python
def get_important_features(df):
    important_features= []
    for i in range(0,df.shape[0]):
        important_features.append(df['title'][i]+df['company_name'][i] + df['location'][i]+
                                  df['description'][i] + df['schedule_type'][i]+df['description_tokens'][i])
    return important_features
    
```

```python
# Now we clean our columns by removing punctuations
import string 
from nltk.corpus import stopwords
```

```python
def normalise_text (text):
    text = text.lower() # lowercase
    text = text.replace(r"\#","") # replaces hashtags
    text = text.replace(r"http\S+","URL")  # remove URL addresses
    text = text.replace(r"@","")
    text = text.replace(r"[^A-Z .,a-z0-9()!?\'\`\"]", " ")
    text = text.replace("\s{2,}", " ")
      # Check characters to see if they are in punctuation
    text = [char for char in text if char not in string.punctuation]

    # Join the characters again to form the string.
    text = ''.join(text)
    
    # Now just remove any stopwords
    return [word for word in text.split() if word.lower() not in stopwords.words('english')]
```

```python
def text_process(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    # Check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]

    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    
    # Now just remove any stopwords
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df.head()
```

```python
tfidf = TfidfVectorizer()
```

```python
# Recommending jobs based on some features for those interested in swiching jobs
```

```python
relevant = tfidf.fit_transform(df['important_features'])
c_similar = cosine_similarity(relevant)

indices = pd.Series(df.index, index=df['title'])


def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(c_similar[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:10]
    job_indices = [i[0] for i in sim_scores]
    print('The following jobs might interest you')

    return df['title'].iloc[job_indices]
```
