Using The logic above I will create a function that compares the temperature changes for each month for each year and plots them side by side

```python
def year_comparison(year1,year2,Months,df=df):
    """This function compares the temperature changes for any two years and plots them
    Args:
    year1,year2(str):These are the years we intend to study for instance Y2018 & Y2019
    The Years range from 1961 to 2019
    Returns :
    Plots 
    """
    fig = plt.figure(figsize=(20,8))
    fig.suptitle(f"The comparison of {year1}and {year2}")
    plt.subplot(2,2,1)
    sns.lineplot(x=df[Months],y=df[year1],color='black')
    plt.title(f"Temperature changes for  {year1}")
    plt.subplot(2,2,2)
    sns.lineplot(y=df[year2],x=df[Months],color='yellow')
    plt.grid()
    plt.title(f"Year Changes in temperatures for {year2}")
    
```

```python
year_comparison('Y2015','Y1998','Months')
```

```python
# 1961  and 2019
year_comparison('Y1961','Y2019','Months')
```

```python
year_comparison('Y2000','Y2019','Months')
```

Temperatures have spiked between the two Years i/e 1961 and 2019 ,there has been a clear global warming on large scale,February and December have the highest temperatures
